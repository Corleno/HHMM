```python
import graphviz as gv

def add_nodes(graph, nodes):
    for n in nodes:
        if isinstance(n, tuple):
            graph.node(n[0], **n[1])
        else:
            graph.node(n)
    return graph

def add_edges(graph, edges):
    for e in edges:
        if isinstance(e[0], tuple):
            graph.edge(*e[0], **e[1])
        else:
            graph.edge(*e)
    return graph


graph = gv.Digraph()
# graph.attr(rankdir="LR")

graph =  add_nodes(graph, [
        ('A', {'label': 'Normal'}),
        ('B', {'label': 'Low grade'}),
        ('C', {'label': 'Death'})
    ])

graph.edge('A', 'B', constraint='false')
graph.edge('B', 'A', constraint='false')
graph.edge('B', 'A', constraint='false', style = "dashed")
graph.edge('A', 'C')
graph.edge('B', 'C')



print (graph)
graph.render('M0', view=True)
```

```python
import graphviz as gv

def add_nodes(graph, nodes):
    for n in nodes:
        if isinstance(n, tuple):
            graph.node(n[0], **n[1])
        else:
            graph.node(n)
    return graph

def add_edges(graph, edges):
    for e in edges:
        if isinstance(e[0], tuple):
            graph.edge(*e[0], **e[1])
        else:
            graph.edge(*e)
    return graph


graph = gv.Digraph()
# graph.attr(rankdir="LR")

subgraph0 = gv.Digraph('cluster_0')
subgraph1 = gv.Digraph('cluster_1')


graph =  add_nodes(graph, [
        ('gamma', {'label': '<&#947;>'}),
        ('beta', {'label': '<&#946;>'}),
        ('lambda', {'label': '<&#955;>'})
])

subgraph0.attr(style='solid', label = 'N', labeljust = 'r', labelloc = 'b')
subgraph0 =  add_nodes(subgraph0, [
        ('z', {'label': '<z<SUB>i</SUB>>'}),
        ('y', {'label': '<y<SUB>i</SUB>>', 'style': 'filled'})
])
subgraph0.edge('z', 'y')



subgraph1.attr(style = 'solid', label = '<&#x221e;>', labeljust = 'r', labelloc = 'b')
subgraph1 =  add_nodes(subgraph1, [
        ('theta', {'label': '<&#952;<SUB>k</SUB>>'})
])


graph.subgraph(subgraph0)
graph.subgraph(subgraph1)

graph.edge('gamma', 'beta')
graph.edge('beta', 'z')
graph.edge('lambda', 'theta')
graph.edge('theta', 'y')
print (graph)
graph.render('DP0', view=True)
```

```python
graph = gv.Digraph()
# graph.attr(rankdir="LR")

subgraph0 = gv.Digraph('cluster_0')
subgraph1 = gv.Digraph('cluster_1')
subgraph2 = gv.Digraph('cluster_2')


graph =  add_nodes(graph, [
        ('gamma', {'label': '<&#947;>'}),
        ('beta', {'label': '<&#946;>'}),
        ('lambda', {'label': '<&#955;>'}),
        ('alpha', {'label': '<&#945;>'})
])

subgraph0.attr(style='solid', label = 'J', labeljust = 'r', labelloc = 'b')
subgraph0 =  add_nodes(subgraph0, [
        ('pi_j', {'label': '<&#960;<SUB>j</SUB>>'}),
])
subgraph0.edge('pi_j', 'z_ji')

subgraph2.attr(style='solid', label = '<N<SUB>ij</SUB>>', labeljust = 'r', labelloc = 'b')
subgraph2 = add_nodes(subgraph2, [
    ('z_ji', {'label': '<z<SUB>ij</SUB>>'}),
    ('y_ji', {'label': '<y<SUB>ij</SUB>>', 'style': "filled"})
])
subgraph2.edge('z_ji', 'y_ji')

subgraph1.attr(style = 'solid', label = '<&#x221e;>', labeljust = 'r', labelloc = 'b')
subgraph1 =  add_nodes(subgraph1, [
        ('theta', {'label': '<&#952;<SUB>k</SUB>>'})
])

subgraph0.subgraph(subgraph2)
graph.subgraph(subgraph0)
graph.subgraph(subgraph1)


graph.edge('gamma', 'beta')
graph.edge('alpha', 'pi_j')
graph.edge('lambda', 'theta')
graph.edge('beta', 'pi_j')
graph.edge('theta', 'y_ji')
print (graph)
graph.render('HDP0', view=True)
```
